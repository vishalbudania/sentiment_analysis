```python
# EXTRACT THE NEWS INFORMATION
options = webdriver.ChromeOptions()
driver = webdriver.Chrome("/home/kira/Documents/chromedriver",options=options)
driver.get('https://www.google.com/search?q=jet+airways&tbm=fin#scso=_rrANXZDJCpLc9QPH5Z7QCA2:0')
html_source=driver.page_source
soup=BeautifulSoup(html_source,'lxml')
tables=soup.find_all('div',class_='dbsr')
titles=soup.find_all('div',class_="nDgy9d")
srcs=soup.find_all('span',class_="wqg8ad")
dtitles = []
dsrc = []
urls = []
for i in range(0,len(tables)):
    tbl = tables[i].find('a') 
    src = srcs[i].text
    title = titles[i].text 
    link = tbl.get('href') 
    dtitles.append(title)
    dsrc.append(src)
    urls.append(link)
        
df = pd.DataFrame()
df['Title'] = dtitles
df['Src'] = dsrc
df['URL'] = urls
df_stock=df[:5]
```
